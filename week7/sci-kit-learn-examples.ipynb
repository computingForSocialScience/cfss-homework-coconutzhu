# Machine Learning with python's scikit-learn

### the vector word model

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = CountVectorizer(min_df=1)
```

```python
vectorizer
```

```python
corpus = ['This is the most important document', \
          'This is the second most important documents',\
          'And the third one, only one',\
          'is this the most important document?']
corpus
```

```python
X = vectorizer.fit_transform(corpus)
vectorizer.get_feature_names()
```

```python
X.toarray()
```

```python
#note that this new document doesn't have ANY of the features from the previous one
vectorizer.transform(["now for something completely different"]).toarray()
```

```python
bigram_vectorizer = CountVectorizer(ngram_range=(1,2), token_pattern=r'\b\w+\b')
analyze = bigram_vectorizer.build_analyzer()
analyze('Bi-grams are awesome!')==(['bi','grams','are','awesome','bi grams', 'grams are', 'are awesome'])
```

### Calculating distance between word vectors

```python
from scipy.spatial import distance
from itertools import combinations

#the distance module in scipy has many vector distance options
#the combinations function takes every unique combination of elements from an iterable python object

i=1
comparisons = combinations(range(4), 2)
for compare in comparisons:
    print "Comparison %d" % i 
    print "Doc1:", corpus[compare[0]], "; Doc2:", corpus[compare[1]]
    print "Cosine distance:", distance.cosine(X[compare[0]].toarray(), X[compare[1]].toarray())
    print "Euclidean distance:", distance.euclidean(X[compare[0]].toarray(), X[compare[1]].toarray())
    print "Manhattan distance:", distance.cityblock(X[compare[0]].toarray(), X[compare[1]].toarray())
    i+=1
```

### Weighting: the tf.idf weighting scheme

```python
from sklearn.feature_extraction.text import TfidfTransformer
transformer=TfidfTransformer(sublinear_tf=True)
transformer
```

```python
#this is for the corpus example above
tfidf = transformer.fit_transform(X)
for x in range(4):
    print corpus[x]
    print tfidf.toarray()[x]
```

##### A more substantial example: on a dataset of press releases from Senators Clinton, Lott and Obama

```python
import os

DIR = "/Users/jevans/Documents/Classes/Content Analysis/press_releases"
press_releases_clinton = [open(os.path.join(DIR, "clinton", f)).read() for f in os.listdir(os.path.join(DIR, "clinton"))]
press_releases_obama = [open(os.path.join(DIR, "obama", f)).read() for f in os.listdir(os.path.join(DIR, "obama"))]
press_releases_lott = [open(os.path.join(DIR, "lott", f)).read() for f in os.listdir(os.path.join(DIR, "lott"))]
press_releases = press_releases_clinton + press_releases_obama + press_releases_lott

#detailed labels
press_releases_clintonL = [f for f in os.listdir(os.path.join(DIR, "clinton"))]
press_releases_obamaL = [f for f in os.listdir(os.path.join(DIR, "obama"))]
press_releases_lottL = [f for f in os.listdir(os.path.join(DIR, "lott"))]
press_releasesL = press_releases_clintonL + press_releases_obamaL + press_releases_lottL
```

```python
print press_releases_clinton[0]
```

enter class labels:

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
labels = ["clinton"]*len(press_releases_clinton) + ["obama"]*len(press_releases_obama) + ["lott"]*len(press_releases_lott) 
le.fit(labels)
le.classes_
```

```python
limiting_vectorizer = CountVectorizer(min_df=3, max_df=100)
releases = limiting_vectorizer.fit_transform(press_releases)
tfidf_release = transformer.fit_transform(releases)
```

```python
km_releases = KMeans(n_clusters=3, init='k-means++', max_iter=100, n_init=1)
km_releases.fit(tfidf_release)
for x in range(len(press_releases)):
    print "Cluster: %d" % km_releases.labels_[x], labels[x] , press_releasesL[x]
```

```python
from sklearn import metrics
metrics.confusion_matrix(km_releases.labels_, le.transform(labels))
```

### Classification example

```python
from sklearn import svm
X = tfidf_release[20:,:]
y = le.transform(labels)[20:]
clf = svm.SVC(probability=True)

#run on all but the first 20
clf.fit(X,y)
```

```python
clf.score(X,y)
```

Predict on some other data set (in this case the first 20 of this dataset) with the same number of features
